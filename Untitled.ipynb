```python
import pandas as pd
import networkx as nx

dades_0 = pd.read_json("data/datathon_participants.json")
dades = dades_0.drop(['email', 'shirt_size', 'dietary_restrictions', 'interests', 'introduction', 'future_excitement', 'age'
                      'fun_fact', 'interest_in_challenges', 'programming_skills', 'experience_level', 'hackathons_done', 'objective', 'technical_project', 'preferred_role' ], axis = 1)

"""sols"""
sols = dades[dades['preferred_team_size'] == 1]  # tenim ja els que volen anar sols

# Inicialitzar el diccionari si no existeix
tancat = {}

# Crear subtaules per a cada participant a 'sols'
for i, row in sols.iterrows():
    subtaula = pd.DataFrame([row])
    tancat[f"subtaula_{len(tancat) + 1}"] = subtaula

"""amics"""
# Convertir a DataFrame
df = pd.DataFrame(dades)

# Construir el grafo
G2 = nx.Graph()

# Afegir nodes i arestes
for _, row in df.iterrows():
    persona = row['id']
    amics = row['friend_registration']
    for amic in amics:
        G2.add_edge(persona, amic)

# Trobar components connectats
components = list(nx.connected_components(G2))

pendents_amics = {}

# Crear subtaules per a cada component i emmagatzemar-les amb noms específics
for i, component in enumerate(components, 1):
    subtaula = df[df['id'].isin(component)].reset_index(drop=True)  # Reiniciar índexos

    if subtaula.shape[0] > 1 and subtaula.iloc[1]['preferred_team_size'] == subtaula.shape[0]:
        tancat[f"subtaula_{len(tancat) + 1}"] = subtaula
    else:
        pendents_amics[f"subtaula_{len(pendents_amics) + 1}"] = subtaula

"""sense amics"""

pendents_NOamics = {}

sense_amics = df[df['friend_registration'].apply(lambda x: len(x) == 0)]


"""nivell"""


# Crear las categorías de estudio
levels = ["1st year", "2nd year", "3rd year", "4th year", "Masters", "PhD"]


# Paso 1: Separar por nivel y almacenar en 'pendents'
for level in levels:
    # Filtrar las filas para la categoría actual
    level_data = sense_amics[sense_amics['year_of_study'] == level]
    
    # Almacenar las subtablas por nivel
    pendents_NOamics[level] = level_data



# Crear las universidades

pendents_NOamics_unis = {}
universidades = ["Universitat Polit\u00e8cnica de Catalunya (UPC)", 
                 "Universitat Aut\u00f2noma de Barcelona (UAB)", 
                 "Universitat de Barcelona (UB)", 
                 "Universitat Internacional de Catalunya (UIC)"]

"""uni"""
len1 = len(pendents_NOamics)
# Paso 2: Separar por universidad dentro de cada nivel
for level in pendents_NOamics.items():
    # Crear subtablas para cada universidad en la categoría de nivel
    for uni in universidades:
        if uni == "Universitat Polit\u00e8cnica de Catalunya (UPC)":
            clave = f"{level}_UPC"
        elif uni == "Universitat Aut\u00f2noma de Barcelona (UAB)": 
            clave = f"{level}_UAB"
        elif uni == "Universitat de Barcelona (UB)": 
            clave = f"{level}_UB"
        elif uni == "Universitat Internacional de Catalunya (UIC)":
            clave = f"{level}_UIC"
        
        # Filtrar los datos por universidad y almacenar la subtabla
        pendents_NOamics_unis[clave] = pendents_NOamics[level][pendents_NOamics[level]['university'] == uni]
    
    # Crear una subtabla para "otras universidades" si es necesario
    clave_otra = f"{level}_otras"
    pendents_NOamics_unis[clave_otra] = level_data[~level_data['university'].isin(universidades)]

    # Eliminar la subtabla por nivel una vez que se ha dividido por universidad
    pendents_NOamics = pendents_NOamics_unis

print(pendents_NOamics)    
```
